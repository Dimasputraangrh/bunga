```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

LOAD DATASET

```python
df = pd.read_csv("/content/youtube-top-100-songs-2025 (1).csv")
df.head(25)
```

```python
df['view_count'] = df['view_count'].astype(float)
```

```python
df['view_count'].isna().sum()
```

FITUR NUMERIK

```python
df['title_length'] = df['title'].str.len()

X = df[['title_length']]
y = df['view_count']
```

```python
y = np.log1p(y)
```

SPLIT DATA

```python
X_train, X_test, y_train, y_test, title_train, title_test = train_test_split(
    X, y, df['title'],
    test_size=0.2,
    random_state=42
)
```

NORMALISASI DATA

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

BUAT MODEL REGRESI

```python
model = Sequential([
    Dense(32, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(16, activation='relu'),
    Dense(1)
])

model.compile(
    optimizer='adam',
    loss='mse'
)
```

TRAIN MODEL

```python
model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=8,
    verbose=1
)
```

PREDIKSI

```python
y_pred_log = model.predict(X_test)

y_pred = np.expm1(y_pred_log).flatten()
y_test_real = np.expm1(y_test)
```

```python
def to_juta(x):
    return f"{x/1_000_000:.0f} jt"
```

HASIL

```python
hasil_prediksi = pd.DataFrame({
    'Judul Video YouTube': title_test.values,
    'Actual View': [to_juta(v) for v in y_test_real],
    'Predicted View': [to_juta(v) for v in y_pred]
})
hasil_prediksi['Actual_View_Num'] = (
    hasil_prediksi['Actual View']
    .str.replace(' jt', '', regex=False)
    .astype(float)
)

hasil_prediksi = hasil_prediksi.sort_values(
    by='Actual_View_Num',
    ascending=False
)

hasil_prediksi = hasil_prediksi.drop(columns=['Actual_View_Num'])
```
