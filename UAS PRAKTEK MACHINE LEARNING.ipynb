```python
# Import library
import pandas as pd
import numpy as np

# Load dataset iris.csv
df = pd.read_csv('/content/iris.csv')

# Tampilkan 5 data awal
df.head(50)
```

DATA & LABEL

```python
print("Jumlah data :", df.shape[0])
print("Jumlah fitur:", df.shape[1] - 1)  # dikurangi target
X = df.drop(columns='virginica') # Label Spesies
y = df['virginica']
```

distrubusi

```python
print("Distribusi kelas target:")
print(df['virginica'].value_counts())
```

REPROCESSING BAGI DATA

```python
X = df.drop(columns='virginica')   # ⬅️ ganti kalau nama target beda
y = df['virginica']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

print("Train:", X_train.shape)
print("Test :", X_test.shape)
```

FEATUR SCALING

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

IMPLEMENTASI KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
```

LATIH MODEL

```python
knn.fit(X_train_scaled, y_train)

y_pred = knn.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1-score :", f1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Create a DataFrame for the metrics
metrics_data = {
    'Metric': ['Accuracy', 'Precision', 'Recall', 'F1-score'],
    'Score': [accuracy, precision, recall, f1]
}
metrics_df = pd.DataFrame(metrics_data)

# Create the bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Metric', y='Score', data=metrics_df, palette='viridis')
plt.ylim(0, 1) # Scores are between 0 and 1
plt.title('Performance Metrics of KNN Classifier')
plt.ylabel('Score')

# Add the score values on top of the bars
for index, row in metrics_df.iterrows():
    plt.text(row.name, row.Score + 0.02, round(row.Score, 3), color='black', ha="center")

plt.show()
```
